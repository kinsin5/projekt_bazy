```python
import pandas as pd
```

```python
df = pd.read_csv('Dane - raport walutowy/NR3_EURUSD60.csv')
```

```python
df['datetime_str'] = df['date'] + ' ' + df['time']
df['datetime'] = pd.to_datetime(df['datetime_str'], format='%Y.%m.%d %H:%M')
```

```python
df = df.drop(columns=['date', 'time', 'datetime_str'])
```

```python
df
```

```python
from sqlalchemy import create_engine

DB_PASS = 'postgres'

# Create engine (replace with your own DB credentials)
engine = create_engine(f'postgresql+psycopg2://postgres:{DB_PASS}@localhost:5432/currencies')

# Insert into table 'currencies'
df.to_sql('currencies', engine, if_exists='append', index=False)
```
